```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.title('Valor do combustível em dolar')

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany']

for country in gas:
  if country in list_country:
    plt.plot(gas['Year'], gas[country], label= [country], marker = '.')

plt.xticks(gas['Year'][::2])

plt.xlabel('Ano')
plt.ylabel('Dolar')

plt.legend()
plt.savefig('Country.pdf')
plt.show()

```
